#Machine Learning Model Creation

##Naive Bayes Model

```python
from scipy.stats import multivariate_normal as mvn
```

```python
class GaussNB():

    def fit(self, X, y, epsilon=1e-3):

        self.likelihoods = dict()
        self.prior = dict()

        self.K = set(y.astype(int))

        for k in self.K:
            X_k = X[y==k, :]
            self.likelihoods[k]= {"mean":X_k.mean(axis=0), "cov": X_k.var(axis=0) + epsilon}
            #print(self.likelihoods)
            self.prior[k]=len(X_k)/len(X)
            #print(self.prior)

    def predict(self, X):
        N,D = X.shape
        P_hat=np.zeros((N, len(self.K)))

        for k, l in self.likelihoods.items():
            P_hat[:,k]=mvn.logpdf(X,l["mean"], l["cov"])+np.log(self.prior[k])
        return P_hat.argmax(axis=1)
```

```python
def accuracy (y,y_hat):
    return np.mean(y==y_hat)
```

```python
gnb = GaussNB()
```

```python
gnb.fit(X,y)
```

```python
y_hat = gnb.predict(X)
```

###Train Dataset Accuracy

```python
accuracy(y,y_hat)
```

```python
y_hat_test = gnb.predict(X_test)
```

###Test Dataset Accuracy

```python
accuracy(y_test,y_hat_test)
```

##Non-Naive Bayes Model

```python
class GaussBayes():
  def fit(self, X, y, epsilon =1e-3):
    self.likelihoods = dict()
    self.priors = dict()
    self.K = set(y.astype(int))

    for k in self.K:
      X_k = X[y==k,:]
      N_k,D = X_k.shape
      mu_k = X_k.mean(axis=0)

      self.likelihoods[k] = {"mean": X_k.mean(axis=0),
                             "cov":(1/(N_k-1))*np.matmul((X_k-mu_k).T,X_k-mu_k)+epsilon*np.identity(D)}

      self.priors[k]=len(X_k)/len(X)
  def predict(self, X):
    N,D = X.shape
    p_hat = np.zeros((N, len(self.K)))

    for k, l in self.likelihoods.items():
      p_hat[:, k] = mvn.logpdf(X, l["mean"],l["cov"])+np.log(self.priors[k])
    return p_hat.argmax(axis=1)
```

```python
gbays_non_navie = GaussBayes()
```

```python
gbays_non_navie.fit(X,y)
```

```python
y_hat2_gbayes=gbays_non_navie.predict(X)
```

##KNN

```python
class KNNClassifier():

    def fit(self, X, y):
        self.X = X
        self.y = y

    def predict(self, X, K, epsilon=1e-3):

        N = len(X)
        y_hat = np.zeros(N)

        for i in range(N):
            dist2 = np.sum((self.X-X[i])**2, axis=1)
            idxt = np.argsort(dist2)[:K]
            gamma_k = 1/(np.sqrt(dist2[idxt] + epsilon))

            y_hat[i] = np.bincount(self.y[idxt], weights=gamma_k).argmax()
        return y_hat
```

```python
knn_instance = KNNClassifier()
knn_instance.fit(X,y)
```

```python
y_hat_knn = knn_instance.predict(X, K=20)
```

```python
accuracy(y,y_hat_knn)
```

```python
y_hat_knn_test = knn_instance.predict(X_test, K=20)
```
